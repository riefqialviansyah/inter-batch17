## Creating DataFrame from dictionary

```python
df2 = {'col_1': [3,4,5,0], 'col_2' : ['a','b','c','d']}
df0 = pd.DataFrame(df2)
df0
```

Specify orient='index' to create the DataFrame using dictionary keys as rows:

```python
df3 = pd.DataFrame.from_dict(df2, orient='index')
df3
```

When using the ‘index’ orientation, the column names can be specified manually:

```python
df3 = pd.DataFrame.from_dict(df2, orient='index', columns=['A','B','C','D'])
df3
```

## 4. Conditionally adding column

```python
import numpy as np
```

```python
data["DISTANCE"] = np.where(data["MILES*"] > 5, "Long Trip", "Short Trip")
```

```python
data.head()
```

```python
data["DISTANCE"].value_counts()
```

```python
data["YEAR"] = np.array("2020")
data
```

```python
data["TIME_CAT"] = np.where(data["START_DATE*"].dt.month >= 2, "NEW TRIP", "OLD TRIP")
```

```python
data.head(80)
```

```python
data["TIME_CAT_2"] = np.where(data["START_DATE*"] > '2016-01-05', "NEW TRIP", "OLD TRIP")
```
